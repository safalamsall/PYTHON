#Calculation of outliers

```python
import matplotlib.pyplot as plt
fig, axs = plt.subplots(len([col for col in df.columns if df[col].dtype != 'object']), 3, figsize=(15, 20 ))
row_idx = 0
for i in df.columns:
    if df[i].dtype != 'object':
        q1 = df[i].quantile(0.25)
        q3 = df[i].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr
        outliers = df[(df[i] < lower_bound) | (df[i] > upper_bound)]
        print(f"Total number of outliers in {i} : {len(outliers)}")
        # Plotting the boxplot
        axs[row_idx, 0].boxplot(df[i])
        axs[row_idx, 0].set_title(f'Boxplot of {i}')

        # Plotting the outliers as scatter plot
        axs[row_idx, 1].scatter(df.index, df[i], label='Data')
        axs[row_idx, 1].scatter(outliers.index, outliers[i], color='r', label='Outliers')
        axs[row_idx, 1].legend()
        axs[row_idx, 1].set_title(f'Scatter plot of {i} with Outliers')
        # Move to the next row
        row_idx += 1
# Adjust layout
plt.tight_layout()
plt.show()
```
